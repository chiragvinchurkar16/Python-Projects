```python
passmarks = 40

# creating a new column pass_math, this column will tell us whether the students are pass or fail
s['pass_math'] = np.where(s['MathScore']< passmarks, 'Fail', 'Pass')
m = s['pass_math'].value_counts().plot.bar(color = 'red', figsize = (5, 5))
m.bar_label(m.containers[0])
plt.title('Comparison of students passed or failed in maths')
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

```python
s['pass_reading'] = np.where(s['ReadingScore']< passmarks, 'Fail', 'Pass')
r = s['pass_reading'].value_counts().plot.bar(color = 'blue', figsize = (5, 5))
r.bar_label(r.containers[0])
plt.title('Comparison of students passed or failed in Reading')
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

```python
s['pass_writing'] = np.where(s['WritingScore']< passmarks, 'Fail', 'Pass')
w = s['pass_writing'].value_counts().plot.bar(color = 'green', figsize = (5, 5))
w.bar_label(w.containers[0])
plt.title('Comparison of students passed or failed in writing')
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

```python
#Total score for each students
s['total_score'] = s['MathScore'] + s['ReadingScore'] + s['WritingScore']
s['total_score'].value_counts()
```

```python
s = s.drop(['Unnamed: 0'],axis = 1)
```

```python
#Percentage for each students
s['percentage'] = s['total_score']/3
```

```python
s['status'] = s.apply(lambda x : 'Fail' if x['pass_math'] == 'Fail' or 
                           x['pass_reading'] == 'Fail' or x['pass_writing'] == 'Fail'
                           else 'pass', axis = 1)

st = s['status'].value_counts().plot.bar(color = 'brown', figsize = (5, 5))
st.bar_label(st.containers[0])
plt.title('overall results')
plt.xlabel('status')
plt.ylabel('count')
plt.show()     

#from this graph we conclude which student is fail overall
```

```python
# Assigning grades to the students according to their marks  :
# 0  - 40 marks : grade F
# 41 - 60 marks : grade E
# 60 - 70 marks : grade D
# 70 - 80 marks : grade C
# 80 - 90 marks : grade B
# 90 - 100 marks : grade A

def getgrade(percentage, status):
  if status == 'Fail':
    return 'F'
  if(percentage >= 90):
    return 'A'
  if(percentage >= 80):
    return 'B'
  if(percentage >= 70):
    return 'C'
  if(percentage >= 60):
    return 'D'
  if(percentage >= 40):
    return 'E'
  else :
    return 'F'

s['grades'] = s.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

s['grades'].value_counts()
```

```python
plt.figure(figsize=(10,5))
g = sns.countplot(x = s['grades'], data = s, hue = s['Gender'])
g.bar_label(g.containers[0])
g.bar_label(g.containers[1])
plt.show()

#from this graph we conclude the distribution of grades between males and females
```

```python
sns.countplot(x = s['ParentEduc'], data = s, hue = s['grades'])
plt.xticks(rotation = 75)
plt.grid()
plt.show()
```
